```python
#2 visualize distributions
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
```

```python
# Exponential distribution
plt.hist(np.random.exponential(scale=1, size=100000), 
         density=True, bins=100, histtype="step", 
         color="red", label="Exponential")
```

```python
# Uniform distribution
plt.hist(np.random.rand(100000), 
         density=True, bins=100, histtype="step", 
         color="blue", label="Uniform")
```

```python
# Normal distribution
plt.hist(np.random.randn(100000), 
         density=True, bins=100, histtype="step", 
         color="green", label="Normal")
```

```python
plt.axis([-2.5, 5, 0, 1.1])
```

```python
plt.legend("upper left")
```

```python
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
```

```python
# 3D plot of Z = X^2 + Y^2
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
# Create 3D plot
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
```

```python
ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
